```python
# --- FRUCTIFAI Agri-Wealth AI: Live Google Sheet Version ---
# Predicts farm income using your actual Google Sheet data.

# STEP 1: Install dependencies
!pip install prophet pandas matplotlib openpyxl --quiet

# STEP 2: Import libraries
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# STEP 3: Connect to your Google Sheet
# ‚ö†Ô∏è Replace the link below with your export link
SHEET_URL = "https://docs.google.com/spreadsheets/d/1MwKD-mdGBHXbEN6ZYTKDqAt80UmAfnoyxLoDGugje4M/export?format=xlsx"

# STEP 4: Load your data from Google Sheets
try:
    data = pd.read_excel(SHEET_URL)
    print("‚úÖ Successfully loaded Google Sheet data!\n")
except Exception as e:
    print("‚ö†Ô∏è Error loading data:", e)

print("Data preview:")
print(data.head())

# STEP 5: Compute profit (yield √ó price)
data['profit'] = data['yield_kg'] * data['price_usd']

# STEP 6: Prepare data for Prophet
df = data[['date', 'profit']].rename(columns={'date': 'ds', 'profit': 'y'})

# STEP 7: Train Prophet model
model = Prophet()
model.fit(df)

# STEP 8: Forecast next 30 days
future = model.make_future_dataframe(periods=30)
forecast = model.predict(future)

# STEP 9: Plot forecast
fig = model.plot(forecast)
plt.title("FRUCTIFAI Predicted Farm Income (Next 30 Days)")
plt.xlabel("Date")
plt.ylabel("Profit ($)")
plt.show()

# STEP 10: Display predicted income
predicted_income = forecast.tail(30)['yhat'].sum()
print(f"\nüåæ Predicted total farm income for next 30 days: ${predicted_income:,.2f}")
import smtplib
from email.message import EmailMessage

EMAIL_ADDRESS = "your_email@gmail.com"
EMAIL_PASSWORD = "YOUR_APP_PASSWORD"
recipient_email = "farmer@example.com"

msg = EmailMessage()
msg.set_content(f"FRUCTIFAI Forecast:\nPredicted income next 30 days = ${predicted_income:,.2f}")
msg['Subject'] = "FRUCTIFAI Daily Forecast"
msg['From'] = EMAIL_ADDRESS
msg['To'] = recipient_email

with smtplib.SMTP_SSL('smtp.gmail.com', 465) as smtp:
    smtp.login(EMAIL_ADDRESS, EMAIL_PASSWORD)
    smtp.send_message(msg)

print("üìß Forecast sent successfully!")
```
